```python
import pandas as pd

# Load the file, replace encoding as needed
df = pd.read_csv('cleansuperstore.csv', encoding='latin1')

# Remove rows where ALL cells are empty
df.dropna(how='all', inplace=True)

# Replace empty strings with NaN, then drop rows with NaN
df.replace('', pd.NA, inplace=True)
df.dropna(how='any', inplace=True)  # or 'any'/'all' depending on need

# Save cleaned file
df.to_csv('cleaned_file.csv', index=False)
```
